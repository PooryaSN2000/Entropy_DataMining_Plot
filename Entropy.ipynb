1. Add Library :

```python
from scipy.stats import entropy
import numpy as np, numpy.random
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

2. Example entropy library

```python
entropy([1/2,1/2,1/3], base=2)
```

3. make random dataset of 3 classes:

```python
data = []
for i in range(0,1000):
    data.append(np.random.dirichlet(np.ones(3),size=1))
```

4. Calclute entropy each one :

```python
entropyList = []
for item in data:
    entropyList.append(entropy(item[0], base=2))
```

5. p1 and p2 and p3 :

```python
p1 = []
p2 = []
p3 = []
for x in range(len(data)):
    p1.append(data[x][0][0])
    p2.append(data[x][0][1])
    p3.append(data[x][0][2])
```

6. plot p1 :

```python
plt.plot(p1, entropyList, 'ro') # Your original list
plt.axis([0, 1, 0, 2])              # Modified axis
plt.show()
```

7. plot p2:

```python
plt.plot(p2, entropyList, 'ro') # Your original list
plt.axis([0, 1, 0, 2])              # Modified axis
plt.show()
```

8. plot p3:

```python
plt.plot(p3, entropyList, 'ro') # Your original list
plt.axis([0, 1, 0, 2])              # Modified axis
plt.show()
```

9. plot entropy of p1 p2 p3 :

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x = p1
y = p2
z = p3
c = entropyList

list_name_variables = ['p1', 'p2', 'p3', 'Entorpy'];
ax.set_xlabel('p1')
ax.set_ylabel('p2')
ax.set_zlabel('p3')
img = ax.scatter(x, y, z, c=c, cmap=plt.hot())
fig.colorbar(img)
plt.show()
```

10. other way:

```python
ax = plt.axes(projection='3d')
ax.scatter(p1, p2, p3, c=entropyList, cmap='viridis', linewidth=0.5);
```
